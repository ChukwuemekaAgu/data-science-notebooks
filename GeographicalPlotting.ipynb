```python
fig = go.Figure(
    data=go.Choropleth(
        locations=['AZ','CA','NY'],
        locationmode='USA-states',
        colorscale='Portland',
        text=['text 1','text 2','text 3'],
        z=[1.0,2.0,3.0],
        colorbar_title='Colorbar Title Goes Here'
    )
)
fig.update_layout(
    title_text='USA Choropleth Map',
    geo_scope='usa',
    template='plotly_white',
    title_x=0.5
)

fig.show()
```

```python
df = pd.DataFrame({
    "state": ['AZ','CA','NY'],
    "value": [1.0,2.0,3.0],
    "text": ['Arizona','California','New York']
})

fig = px.choropleth(
    df,
    locations="state",
    locationmode="USA-states",
    color="value",
    hover_name="text",
    color_continuous_scale="Portland",
    title="USA Choropleth Map"
)

fig.update_layout(
    geo_scope="usa",
    template="plotly_white",
    title_x=0.5
)

fig.show()
```

```python
ds = pd.read_csv('2011_US_AGRI_Exports')
```

```python
ds.head()
```

```python
print(ds.columns)
```

```python
fig = px.choropleth(
    ds,
    locations= "code",
    locationmode="USA-states",
    color="total exports",
    hover_name="state",
    hover_data=["category","total veggies","wheat","corn"],
    color_continuous_scale="Greens",
    title="2011 US Agriculture Exports by State"
)

fig.update_layout(
    geo_scope="usa",
    template="plotly_white",
    title_x=0.5
)

fig.show()
```

```python
fig = px.choropleth(
    ds2,
    locations= "CODE",
    locationmode="ISO-3",
    color="GDP (BILLIONS)",
    hover_name="COUNTRY",
    hover_data=['COUNTRY', 'GDP (BILLIONS)', 'CODE'],
    color_continuous_scale="Greens",
    title="GDP in Billions USD"
)

fig.update_layout(
    geo_scope="world",
    template="plotly_white",
    title_x=0.5
)

fig.show()
```

```python
ex1 = pd.read_csv('2014_World_Power_Consumption')
cc = coco.CountryConverter()
ex1['iso_alpha'] = cc.convert(names=ex1['Country'], to='ISO3')
```

```python
ex1.head()
```

```python
fig = px.choropleth(
    ex1,
    locations= "iso_alpha",
    locationmode="ISO-3",
    color="Power Consumption KWH",
    hover_name='Country',
    hover_data='Power Consumption KWH',
    color_continuous_scale="Greens",
    title="Power Consumption KWH"
)

fig.update_layout(
    geo_scope="world",
    template="plotly_white",
    title_x=0.5
)

fig.show()
```
