```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3
```

```python
fig = px.scatter_3d(
    df3,
    x='x',
    y='y',
    z='z',
    color='z'
)

fig.show()
```

```python
fig = px.scatter_3d(
    df3,
    x='x',
    y='y',
    z='z',
    color='z',
    color_continuous_scale='Viridis',
    labels={
        'x': 'Feature X',
        'y': 'Feature Y',
        'z': 'Target / Response Z'
    },
    title='3D FeatureTarget Relationship'
)

fig.update_traces(
    marker=dict(size=6, opacity=0.85)
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5,
    margin=dict(l=0, r=0, b=0, t=50),
    scene=dict(
        xaxis=dict(backgroundcolor="rgb(245,245,245)"),
        yaxis=dict(backgroundcolor="rgb(245,245,245)"),
        zaxis=dict(backgroundcolor="rgb(245,245,245)")
    )
)

fig.show()
```

```python
fig = px.line(
    df3,
    x='x',
    y='z',
    markers=True,
    labels={'x': 'Feature X', 'z': 'Target Z'},
    title='Effect of Feature X on Target Z'
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5
)

fig.show()
```

```python
fig = px.histogram(df)
fig.show()
```

```python
df['spread'] = df['A'] - df['B']
```

```python
fig = px.line(
    df,
    x=df.index,
    y='spread',
    labels={
        'x': 'Index / Time',
        'spread': 'Spread (A  B)'
    },
    title='Spread Between A and B'
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5
)

fig.add_hline(
    y=0,
    line_dash="dash",
    line_color="gray",
    annotation_text="Zero Spread",
    annotation_position="bottom right"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x='A',
    y='B',
    size=df['C'].abs(),
    color=df['C'],
    size_max=40,
    color_continuous_scale='RdBu',
    color_continuous_midpoint=0,
    labels={
        'A': 'Feature A',
        'B': 'Feature B',
        'C': 'Signed Magnitude (C)'
    },
    title='Bubble Chart: Magnitude and Sign of C'
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5
)

fig.show()
```

```python
fig = px.scatter_matrix(
    df,
    dimensions=df.columns,
    title='Scatter Matrix with Feature Distributions'
)

fig.update_traces(
    diagonal_visible=True,
    showupperhalf=False
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5
)

fig.show()
```
