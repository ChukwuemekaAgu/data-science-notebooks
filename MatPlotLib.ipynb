```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y,'r')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('X and Y Coordinate')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

```python
# OBJECT ORIENTED METHOD
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X and Y Coordinates')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Coordinate')

axes[1].plot(y,x)
axes[1].set_title('Second Coordinate')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y,'r')

axes[1].plot(y,x)

plt.tight_layout()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
import pandas as pd
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('X axis')

ax.set_ylabel('Y axis')

ax.set_title('Matplotlib Exercise')
ax.plot(x,y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color='red')
ax2.plot(x,y)
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

# Large
ax.plot(x,z)
ax.set_xlabel('x index')
ax.set_ylabel('z index')

# Inside
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x index')
ax2.set_ylabel('y index')

ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,color='blue',ls='--')
axes[1].plot(x,z,color='red',lw=3)

fig.tight_layout()
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y,color='purple')
axes[1].plot(x,z,color='red',ls='--')
```
